```python
import os

def replace_labels_in_txt_files(input_dir, old_label, new_label):

    if not os.path.isdir(input_dir):
        print(f"Error: The directory {input_dir} does not exist.")
        return

    for file_name in os.listdir(input_dir):
        if file_name.endswith(".txt"):
            file_path = os.path.join(input_dir, file_name)
            try:
                with open(file_path, 'r') as file:
                    lines = file.readlines()

                updated_lines = []
                for line in lines:
                    parts = line.strip().split()
                    if parts:
                        if parts[0] == old_label:
                            parts[0] = new_label
                        updated_lines.append(" ".join(parts))

                with open(file_path, 'w') as file:
                    file.write("\n".join(updated_lines))
                print(f"Processed file: {file_name}")

            except Exception as e:
                print(f"Error processing file {file_name}: {e}")

```

```python
input_directory = "./datasets/carr.v1i.yolov11/test/labels/"
old_label = "0"
new_label = "2"
replace_labels_in_txt_files(input_directory, old_label, new_label)
```

```python
input_directory = "./datasets/carr.v1i.yolov11/train/labels/"
old_label = "0"
new_label = "2"
replace_labels_in_txt_files(input_directory, old_label, new_label)
```
